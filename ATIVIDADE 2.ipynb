```python
# Selecionar arquivo para análise
import os

# Pegando o caminho do diretório que estou
rootPath = os.getcwd()
print(rootPath)

# Definindo o diretório com os arquivos metar
metSuperficiePath = rootPath +'\\imputs\\meteoSuperficie'
print(metSuperficiePath)

# Listando os arquivos dentro do diretório
files = os.listdir(metSuperficiePath)
print(files)

```

# Perfil de temperatura ( 5 dias no verão)


```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
from siphon.simplewebservice.wyoming import WyomingUpperAir

# Função para calcular o gradiente de temperatura
def calculate_gradient_temperature(temperature, height):
    deltaT = np.diff(temperature)
    deltaZ = np.diff(height)
    dTdZ = (deltaT / deltaZ) * 100  # Gradiente de temperatura a cada 100 m
    return dTdZ

# Função para identificar pontos de inversão
def find_inversion_points(gradient_temperature, height):
    inversion_points = []
    for i in range(1, len(gradient_temperature)):
        if gradient_temperature[i] * gradient_temperature[i - 1] < 0:
            inversion_points.append((height[i - 1] + height[i]) / 2)
    return inversion_points

# Função para plotar o perfil de temperatura e os pontos de inversão
def plot_temperature_profile(temperature, height, inversion_points, date, ax):
    ax.plot(temperature, height, linestyle='-', color='blue')
    ax.set_xlabel('Temperatura (°C)')
    ax.set_ylabel('Altitude (m)')
    ax.set_title(f'Perfil de Temperatura - {date}')
    for point in inversion_points:
        ax.axhline(y=point, color='red', linestyle='--', linewidth=1)
    ax.grid(True)

# Função para plotar o gradiente de temperatura
def plot_temperature_gradient(gradient_temperature, height, date, ax):
    ax.plot(gradient_temperature, height[:-1], linestyle='-', color='green')
    ax.set_xlabel('Gradiente de Temperatura (°C/100m)')
    ax.set_ylabel('Altitude (m)')
    ax.set_title(f'Gradiente de Temperatura - {date}')
    ax.grid(True)

# Data para análise (5 dias no verão)
summer_dates_midnight = [
    datetime(2023, 1, 1, 0),
    datetime(2023, 1, 2, 0),
    datetime(2023, 1, 3, 0),
    datetime(2023, 1, 4, 0),
    datetime(2023, 1, 5, 0)
]

summer_dates_noon = [
    datetime(2023, 1, 1, 12),
    datetime(2023, 1, 2, 12),
    datetime(2023, 1, 3, 12),
    datetime(2023, 1, 4, 12),
    datetime(2023, 1, 5, 12)
]

# Local da análise
station = 'SBFL'

# Loop sobre as datas de verão à meia-noite
for date in summer_dates_midnight:
    # Extraindo os dados
    df = WyomingUpperAir.request_data(date, station)
    # Calculando o gradiente de temperatura
    gradient_temperature = calculate_gradient_temperature(df['temperature'], df['height'])
    # Identificando os pontos de inversão
    inversion_points = find_inversion_points(gradient_temperature, df['height'])
    
    # Criando uma figura com dois subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Plotando o perfil de temperatura e os pontos de inversão no primeiro subplot
    plot_temperature_profile(df['temperature'], df['height'], inversion_points, date.strftime('%Y-%m-%d %H:%M'), ax1)

    # Plotando o gradiente de temperatura no segundo subplot
    plot_temperature_gradient(gradient_temperature, df['height'], date.strftime('%Y-%m-%d %H:%M'), ax2)

    plt.tight_layout()
    plt.show()

# Loop sobre as datas de verão ao meio-dia
for date in summer_dates_noon:
    # Extraindo os dados
    df = WyomingUpperAir.request_data(date, station)
    # Calculando o gradiente de temperatura
    gradient_temperature = calculate_gradient_temperature(df['temperature'], df['height'])
    # Identificando os pontos de inversão
    inversion_points = find_inversion_points(gradient_temperature, df['height'])
    
    # Criando uma figura com dois subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Plotando o perfil de temperatura e os pontos de inversão no primeiro subplot
    plot_temperature_profile(df['temperature'], df['height'], inversion_points, date.strftime('%Y-%m-%d %H:%M'), ax1)

    # Plotando o gradiente de temperatura no segundo subplot
    plot_temperature_gradient(gradient_temperature, df['height'], date.strftime('%Y-%m-%d %H:%M'), ax2)

    plt.tight_layout()
    plt.show()
```

# Albedo, Razão de bowen e Rugosidade

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para o arquivo CSV
caminho_arquivo = r'C:\Users\sah27\Documents\GitHub\ENS_5173_202401\Imputs\EZ_AERMOD_sabrina\landUseCharaterroserrena.csv'

# Abrir o arquivo CSV como um DataFrame do pandas, especificando o encoding
df = pd.read_csv(caminho_arquivo, encoding='latin1')

# Configurações para exibir todas as colunas
pd.set_option('display.max_columns', None)

# Exibir DataFrame
display(df)

# Plotar os gráficos
plt.figure(figsize=(14, 12))

# Plot para rugosidade (z0)
plt.subplot(3, 1, 1)
bar_width = 0.2
sectors = df['Sectors']
indices = range(len(sectors))
plt.barh([i - bar_width for i in indices], df['z0_spring'], height=bar_width, label='Primavera')
plt.barh(indices, df['z0_summer'], height=bar_width, label='Verão')
plt.barh([i + bar_width for i in indices], df['z0_autumn'], height=bar_width, label='Outono')
plt.barh([i + 2 * bar_width for i in indices], df['z0_winter'], height=bar_width, label='Inverno')
plt.yticks(indices, sectors)  # Garantir que os setores sejam exatamente os da tabela
plt.xlabel('Rugosidade (z0)')
plt.title('Rugosidade (z0) por Estação')
plt.legend()

# Plot para albedo
plt.subplot(3, 1, 2)
plt.barh([i - bar_width for i in indices], df['alb_spring'], height=bar_width, label='Primavera')
plt.barh(indices, df['alb_summer'], height=bar_width, label='Verão')
plt.barh([i + bar_width for i in indices], df['alb_autumn'], height=bar_width, label='Outono')
plt.barh([i + 2 * bar_width for i in indices], df['alb_winter'], height=bar_width, label='Inverno')
plt.yticks(indices, sectors)  # Garantir que os setores sejam exatamente os da tabela
plt.xlabel('Albedo')
plt.title('Albedo por Estação')
plt.legend()

# Plot para razão de bo
plt.subplot(3, 1, 3)
plt.barh([i - bar_width for i in indices], df['bo_spring'], height=bar_width, label='Primavera')
plt.barh(indices, df['bo_summer'], height=bar_width, label='Verão')
plt.barh([i + bar_width for i in indices], df['bo_autumn'], height=bar_width, label='Outono')
plt.barh([i + 2 * bar_width for i in indices], df['bo_winter'], height=bar_width, label='Inverno')
plt.yticks(indices, sectors)  # Garantir que os setores sejam exatamente os da tabela
plt.ylabel('Quadrante')
plt.xlabel('Razão de bowen')
plt.title('Razão de Bowen por Estação')
plt.legend()

# Ajustar layout
plt.tight_layout()
plt.show()
```
