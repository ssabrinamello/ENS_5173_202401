```python
#criando números aleatórios e mostrando a dimensão da matriz
b= np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
# criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 # valor mínimo 
vmax = 100 # valor máximo 
c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose()) # adicionado o transpose inverte eixos

```

```python
# Plotando apenas a primeira linha 
x = np.array(range(0,np.size(c[0,:]))) # convertendo o range para 
print(x)
plt.plot(x,c[0,:])
plt.title('Gráfico com valores inteiros no eixo x')
plt.xticks(x) # Definindo os valores do eixo x 
plt.plot(c[0,:]) # plotando os valores da primeira linha todas as colunas

```

```python
# criando um gráfico de barras e exportando 
fig, ax = plt.subplots(3,1) #criou a figura com os eixos. ax são os eixos
ax[0].bar(x,c[:,0]) # bar significa que vai plotar um gáfico de barras
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

#exportar a figura 
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
fig.savefig(figFolder +'Figura01.png') # salvou a figura na pasta
```
